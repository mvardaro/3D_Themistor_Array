```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
nc_files = 'data/deployment0001_RS03ASHS-MJ03B-07-TMPSFA301-streamed-tmpsf_sample_20170904T030020.340942-20170911T030010.192657.nc'
```

```python
ds = xr.open_dataset(nc_files)
ds = ds.swap_dims({'obs': 'time'})
```

```python
ds['temperature04'].plot.line()
plt.ylabel('Temperature in degrees Celsius')
plt.xlabel('Time')
plt.legend()
plt.show()
```
